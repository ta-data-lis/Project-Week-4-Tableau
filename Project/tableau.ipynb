Rename columns:

```python
ds.columns = ["Country", "Sex", "Occupation", "Industry", "Age", "Value"]
```

```python
ds.Value
```

Taking out the all the rows without values.

```python
ds1 = ds[":"!=ds["Value"]]
ds1["Value"] = [int(i) for i in ds1["Value"]]
```

Adding a count just to check for the amount of grouped rows when using groupby() later

```python
ds1["count"]=1
ds1.head(3)
```

Taking out countries that only have summed values and value 0 :

```python
out = ["DK", "ES", "FI", "IS", "IT", "LT", "NL", "NO", "SE"]
ds1["mark"] = [1 if i not in out else 0 for i in ds1["Country"]]
ds1.head()
```

```python
ds1 = ds1[1==ds1["mark"]]
ds1.drop(columns="mark", inplace=True)
ds1.head()
```

Testing for summed values and counts :

```python
test = ds1.groupby(["Country"], as_index=False).agg({"Value":'sum', 'count':'sum'})
test
```

```python
test1 = ds1.groupby(["Occupation", "Industry", "Age"], as_index=False).agg({"Value":'sum', 'count':'sum'})
test1
```

```python
test1["Relative"]= test1['Value']/sum(test1["Value"])
test1.head(5)
```

### Adding Total value per country and relative values over respective country and over relative values over entirety:

```python
work = ds1[ds1['Value']!=0]
work["Relative / all countries"]= work['Value']/sum(work["Value"])
work.head(10)
```

```python
population = work.groupby("Country", as_index = False).agg({"Value":"sum"})
new_work = work.merge(population, on = "Country")
new_work.columns = ["Country", "Sex", "Occupation", "Industry", "Age", "Value", "count", "Relative / all countries","Total_Population"]
new_work.head()
```

```python
new_work["Relative per country"]= new_work["Value"]/new_work["Total_Population"]
new_work.head()
```

### Marking population that will exit the work force within the next years by splitting by age. Younger than 50 are marked as 'Young' otherwise 'Old':

```python
import re
new_work["Young"] = [1 if i.startswith("Y1") or i.startswith("Y3") else 0 for i in new_work["Age"]]
```

```python
new_work["Old"] = [0 if i.startswith("Y1") or i.startswith("Y3") else 1 for i in new_work["Age"]]
new_work.head(4)
```

```python
new_work["Young or Old"] = ["Young" if i.startswith("Y1") or i.startswith("Y3") else "Old" for i in new_work["Age"]]
new_work.head()
```

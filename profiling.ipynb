```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
#this dataset includes all neighborhoods (81) where health unites are located within the city of Vitória. 
#Reminders improve attendance?
#create column 'Condition'
#make waiting time column. make negative values zero (no waiting time)
#|make Handicap binary. Convert 2,3 and 4 to 1
#|Age=-1: make it zero
#convert WaitingTime to number of waited hours

'''
AppointmentDay: the day of the actuall appointment, when they have to visit the doctor.
ScheduledDay: the day someone called or registered the appointment, this is before appointment of course.
WaitingTime = AppointmentDay - ScheduledDay

‘Scholarship’ indicates whether or not the patient is enrolled in Brasilian welfare program Bolsa Família. 
'''
```

```python
#from someone in kaggle

'''
If we interpret the SMS column as an indication of an SMS being sent then it makes sense that 
hospitals would have a rule/system in place to not send SMS to people whose appointments are close 
to the date the appointment was scheduled.
'''
```

```python
medical_appointments = pd.read_csv('medical_appointments_RAW.csv')

medical_appointments.head()
```

```python
#ma = ma1.drop(columns=['Scholarship']) #not relevant
ma = medical_appointments.rename(columns={'Handcap':'Handicap'}) #wording
```

```python
ma.shape #(110527, 13)
```

```python
ma.loc[ma['Age']<0]
```

```python
#Age=-1: make it zero (must be a baby)
ma.loc[ma['Age']<0] = 0
```

```python
ma.loc[ma['Age']<0]
```

```python
ma.Handicap.value_counts()
```

```python
#make Handicap binary. Convert 2,3 and 4 to 1
ma.loc[ma['Handicap']>1] = 1
```

```python
ma.Handicap.value_counts()
```

```python
ma.head()
```

```python
ma.dtypes
```

```python
#convert to datetime type so I can perform operations on it
ma[['ScheduledDay','AppointmentDay']] = ma[['ScheduledDay','AppointmentDay']].apply(pd.to_datetime, utc=True)
```

```python
ma.head()
```

```python
'''
Unfortunately there are no hours for AppointmentDay... would be interesting to investigate wether or not the fact
patients are no-showing is due to appointment hour (e.g. early hours)
'''

#reduce for date only (no hour)
ma['ScheduledDay'] = pd.to_datetime(ma['ScheduledDay']).dt.date
ma['AppointmentDay'] = pd.to_datetime(ma['AppointmentDay']).dt.date
ma.head()
```

```python
#make WaitingTime column (subtracting scheduledDay from appointmentDay)
ma['WaitingTime(days)'] = ma['AppointmentDay'] - ma['ScheduledDay']
```

```python
ma.head()
```

```python
#converting WaitingTime to number of days/int, for plotting purposes later one (tableau)
ma['WaitingTime(days)'] = ma['WaitingTime(days)'] / np.timedelta64(1, 'D')
ma.head()
```

```python
ma_negative = ma.loc[ma['WaitingTime(days)']<0] #all values less then 0 in WaitingTime
ma_negative.shape
```

```python
'''
Negative values in WaitingTime are maybe some error in the original dataset. Doesn't have sense to have negative
time / having an appointment and schedule it after having it.
I decided to drop these values
'''

# Get names of indexes for which column'WaitingTime(days)' has negative value
to_drop = ma_negative.index
 
# Delete these row indexes from dataFrame
ma.drop(to_drop , inplace=True)
```

```python
#converting integers to strings to performace string operations on it
ma[['Hipertension','Diabetes','Alcoholism','Handicap','No-show','Scholarship','SMS_received']] = ma[['Hipertension','Diabetes','Alcoholism','Handicap','No-show','Scholarship','SMS_received']].astype('str') 
```

```python
ma.dtypes
```

```python
#replace of 1's
ma['Hipertension'] = ma['Hipertension'].apply(lambda x: x.replace('1','Hipertension '))
ma['Diabetes'] = ma['Diabetes'].apply(lambda x: x.replace('1','Diabetes '))
ma['Alcoholism'] = ma['Alcoholism'].apply(lambda x: x.replace('1','Alcoholism '))
ma['Handicap'] = ma['Handicap'].apply(lambda x: x.replace('1','Handicap '))
#ma['SMS_received'] = ma['SMS_received'].apply(lambda x: x.replace('1','Yes'))

#to avoid confusion (double negation)
ma['No-show'] = ma['No-show'].apply(lambda x: x.replace('No','Show'))


#replace of 0's
ma['Hipertension'] = ma['Hipertension'].apply(lambda x: x.replace('0',''))
ma['Diabetes'] = ma['Diabetes'].apply(lambda x: x.replace('0',''))
ma['Alcoholism'] = ma['Alcoholism'].apply(lambda x: x.replace('0',''))
ma['Handicap'] = ma['Handicap'].apply(lambda x: x.replace('0',''))
#ma['SMS_received'] = ma['SMS_received'].apply(lambda x: x.replace('0','No'))

#to avoid confusion (double negation)
ma['No-show'] = ma['No-show'].apply(lambda x: x.replace('Yes','No-show'))
```

```python
ma.head()
```

```python
#create columns 'Condition'
ma['Condition'] = ma['Hipertension']+ma['Diabetes']+ma['Alcoholism']+ma['Handicap']
ma.head()
```

```python
ma['Condition'].value_counts()
```

```python
#drop columns
med_apt = ma.drop(columns=['Hipertension','Diabetes','Alcoholism','Handicap'])
med_apt.head()
```

```python
med_apt['Condition'].replace('', 'Other', inplace=True)
```

```python
med_apt['WaitingTime(days)'] = med_apt['WaitingTime(days)'].astype('int')
```

```python
med_apt.dtypes
```

```python
med_apt.head()
```

```python
med_apt.Condition.value_counts()
```

```python
#drop Genders different than F and M
no_gender = med_apt.loc[(med_apt['Gender']==1)|(med_apt['Gender']==0)]
to_drop = no_gender.index
med_apt.drop(to_drop , inplace=True)
```

```python
#new Age column: Age_interval

#babies [0, 2]
med_apt.loc[((med_apt['Age']>=0)&(med_apt['Age']<=2)),'Age_Group'] = 'Babies [0-2]'
#children [3, 10]
med_apt.loc[((med_apt['Age']>=3)&(med_apt['Age']<=10)),'Age_Group'] = 'Children [3-10]'
#teenagers [11, 17]
med_apt.loc[((med_apt['Age']>=11)&(med_apt['Age']<=17)),'Age_Group'] = 'Teenagers [11-17]'
#young adults [18, 30]
med_apt.loc[((med_apt['Age']>=18)&(med_apt['Age']<=40)),'Age_Group'] = 'Young adults [18-40]'
#adults [31, 65]
med_apt.loc[((med_apt['Age']>=41)&(med_apt['Age']<=65)),'Age_Group'] = 'Adults [41-65]'
#elderly [66, 115]
med_apt.loc[((med_apt['Age']>=66)&(med_apt['Age']<=115)),'Age_Group'] = 'Elderly [66-115]'

med_apt.head()
```

```python
#replacement of 1's
med_apt['Scholarship'] = med_apt['Scholarship'].apply(lambda x: x.replace('1','Yes'))
med_apt['SMS_received'] = med_apt['SMS_received'].apply(lambda x: x.replace('1','Yes'))

#replacement of 0's
med_apt['Scholarship'] = med_apt['Scholarship'].apply(lambda x: x.replace('0','No'))
med_apt['SMS_received'] = med_apt['SMS_received'].apply(lambda x: x.replace('0','No'))
```
